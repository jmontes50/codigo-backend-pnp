```python
df_pais = pd.read_csv('Info_pais.csv', encoding="ISO-8859-1", delimiter=';')
```

```python
df_pais
```

```python
df_order = df_pais.sort_values("Esperanza de vida", ascending=True)
```

```python
df_order
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'])
plt.title("Renta per Capita y Esperanza de Vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
```

```python
df_order["Pobl_norm"] = df_order['Poblacion']/(max(df_order['Poblacion'])/10000)
```

```python
df_order.head()
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'],s=df_order['Pobl_norm'])
plt.title("Renta per Capita y Esperanza de Vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'],s=df_order['Pobl_norm'])
plt.title("Renta per Capita y Esperanza de Vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

fig = plt.gcf()
fig.set_size_inches(20, 15)
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'],s=df_order['Pobl_norm'])
plt.title("Renta per Capita y Esperanza de Vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

fig = plt.gcf()
fig.set_size_inches(20, 15)

for i in range(1, 10):
    plt.annotate(df_order["Pais"][i],df_order["Renta per capita"][i], df_order["Esperanza de vida"][i])
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'],s=df_order['Pobl_norm'])
plt.title("Renta per Capita y Esperanza de Vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

fig = plt.gcf()
fig.set_size_inches(20, 15)

for i in range(1, 10):
    plt.annotate(df_order["Pais"][i],(df_order["Renta per capita"][i], df_order["Esperanza de vida"][i]))
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'],s=df_order['Pobl_norm'])
plt.title("Renta per Capita y Esperanza de Vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

fig = plt.gcf()
fig.set_size_inches(20, 15)

for i in range(1, 10):
    plt.annotate(df_order["Pas"][i],df_order["Renta per capita"][i], df_order["Esperanza de vida"][i])
```

```python
plt.scatter(df_order['Renta per capita'], df_order['Esperanza de vida'],s=df_order['Pobl_norm'],c=df_order['Pobl_norm'])
plt.title("Renta per Capita y Esperanza de Vida")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")

fig = plt.gcf()
fig.set_size_inches(20, 15)

#aca lo que hago es obtener el nombre del pad
for i in range(1, 10):
    plt.annotate(df_order["Pas"][i],(df_order["Renta per capita"][i], df_order["Esperanza de vida"][i]))
```
